# ----Separando atributos previsores e classificadores----

```python
previsores = base.iloc[:,0:4].values
pd.DataFrame(previsores)
```

```python
classe = base.iloc[:,4].values
pd.DataFrame(classe)
```

# ----Transformação de variaveis catoricas em numericas----

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
previsores[:,0] = labelencoder.fit_transform(previsores[:,0])
previsores[:,1] = labelencoder.fit_transform(previsores[:,1])
previsores[:,2] = labelencoder.fit_transform(previsores[:,2])
previsores[:,3] = labelencoder.fit_transform(previsores[:,3])
```

```python
pd.DataFrame(previsores)
```

# ----Aprendizagem bayesiana----

```python
from sklearn.naive_bayes import GaussianNB
classificador = GaussianNB()
```

```python
classificador.fit(previsores,classe)
```

### Verificando risco de crédito

```python
# história boa, dívida alta, garantias nenhuma, renda > 35
resultado = classificador.predict([[0,0,1,2]])
resultado
```

```python
# história ruim, dívida alta, garantias adequada, renda < 15
resultado2 = classificador.predict([[3, 0, 0, 0]])
resultado2
```
